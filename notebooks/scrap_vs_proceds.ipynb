# Obtener claves de procedimiento y de UC

```python
proced_format = re.compile('[A-Z]{2}\-[0-9A-Z]{9}\-.*\-[0-9]{4}')

def get_proced(string):
    string=str(string)
    match = proced_format.search(string)
    if match is not None:
        match = match.group(0)
        s = match[0:15]
        if s.find(' ')!=-1:
            s = s.replace(' ', '')
            s = s+match[len(s)+1::]
            s = s.split()[0]
            if len(s)>20:
                return s
            else:
                return None 
        else:
            return match.split()[0]
        
    else: 
        return None
```

```python
scrap['NUMERO_PROCEDIMIENTO'] = scrap.referencia_expediente.map(get_proced)
```
